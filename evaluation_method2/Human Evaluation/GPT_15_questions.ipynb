## Below is GPT3.5 Turbo's answers to 15 selected challenging math questions for human evaluation

```python
import json
import openai
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
def load_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

def chat_gpt(prompt):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content.strip()
```

```python
# Enter API Key
client = openai.OpenAI(
    api_key=""  # Use API key
)

data_path = "test15.json"
data = load_data(data_path)
```

```python
for item in data:
    prompt = f"Solve and explain this math problem: {item['Problem']}"

    response = chat_gpt(prompt)

    print(f"Problem: {item['Problem']}")
    print(f"Options: {item['options']}")
    print(f"Correct option: {item['correct']}")
    print(f"GPT's resonse:{response}")
    print("---------------------------------------------------------------------")
```
