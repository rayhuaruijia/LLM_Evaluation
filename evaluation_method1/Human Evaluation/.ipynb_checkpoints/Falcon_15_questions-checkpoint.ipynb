```python
import json

with open('/content/drive/My Drive/Colab Notebooks/test15.json', 'r') as file:
    data = json.load(file)
```

```python
def chat_falcon(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_length=1024)

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
for item in data:
    prompt = f"Solve and explain this math problem: {item['Problem']} "

    response = chat_falcon(prompt)

    print(f"Problem: {item['Problem']}")
    print(f"Options: {item['options']}")
    print(f"Correct option: {item['correct']}")
    print(f"Falcon's resonse:{response}")
    print("---------------------------------------------------------------------")
```
